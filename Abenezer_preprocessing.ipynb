```python
import cv2
import matplotlib.pyplot as plt 
import os 
import glob

img_dir = r"C:\Users\abeng\OneDrive\Desktop\DIP Project\DIP Project\DIP Project (1)\Scanned_forms"
data_path = os.path.join(img_dir, '*g')
files = glob.glob(data_path)
data = []

for f1 in files:
    img = cv2.imread(f1)
    data.append(img)
```

```python
im = data[24].copy()
```

```python
gray_image = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

```python
img = cv2.medianBlur(gray_image,5)
```

```python
Blurred_image = cv2.adaptiveThreshold(gray_image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,35,2)
```

```python
plt.figure(figsize=(18,17))
plt.imshow(Blurred_image, cmap='gray')
```

```python
ret,thresh = cv2.threshold(Blurred_image,10,255,0)
im_new, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
clone = im.copy()
c = contours
cv2.drawContours(clone, c, -1, (0,255,0), 3)
plt.figure(figsize=(18,17))
plt.imshow(clone, cmap='gray')
```

```python
Image_tobe_Inverted = Blurred_image.copy()
#Image_tobe_cropped = cv2.bitwise_not(Image_tobe_Inverted)
ret,Image_tobe_cropped = cv2.threshold(Image_tobe_Inverted,200,255,cv2.THRESH_BINARY_INV)

index = 0 
for c in contours:
    x,y,w,h = cv2.boundingRect(c)
    if w in range(45,125) and h in range(50,170):
        index+=1
        new_image = Image_tobe_cropped[y:y+h, x:x+w]
        cv2.imwrite(str(index) + '.png', new_image)
cv2.imshow("image",Image_tobe_cropped)
cv2.waitKey(0)
```
